```python
Q2
MongoDB is a popular NoSQL database management system that is widely used for building web applications. Here are some of the key features of MongoDB:

Document-Oriented: MongoDB is a document-oriented database, which means that data is stored in flexible, JSON-like documents that can have nested structures. This allows for more flexible and dynamic data models compared to traditional relational databases.

Scalability: MongoDB is highly scalable, which makes it ideal for handling large volumes of data. It allows you to distribute data across multiple servers and can handle thousands of reads and writes per second.

High Availability: MongoDB provides high availability through replica sets, which are groups of MongoDB servers that maintain identical copies of data. In the event of a server failure, one of the replicas can take over, ensuring that the data remains available.

Indexing: MongoDB supports various types of indexing, including single-field, compound, and geospatial indexes. This allows for faster query performance and efficient data retrieval.

Aggregation: MongoDB provides powerful aggregation capabilities that allow you to perform complex data analysis tasks. The aggregation pipeline framework supports a wide range of operators and stages, enabling you to group, filter, and transform data in real-time.
```

```python
from pymongo import MongoClient

# creating a connection
client = MongoClient('localhost', 27017)

# accessing the database
db = client['mydatabase']

# accessing the collection
col = db['customers']

# querying the collection using find() method
query = { "address": "Park Lane 38" }
result = col.find(query)

# iterating through the result
for record in result:
    print(record)
```

```python
rom pymongo import MongoClient

# creating a connection
client = MongoClient('localhost', 27017)

# accessing the database
db = client['mydatabase']

# accessing the collection
col = db['customers']

# sorting the collection using sort() method
result = col.find().sort("name", 1)

# iterating through the sorted result
for record in result:
    print(record)
```

```python
n MongoDB, there are several methods available for deleting documents from a collection or dropping a collection altogether. Here's a brief explanation of when to use each of the following methods:

delete_one() - This method is used to delete a single document from a collection that matches the specified filter. If there are multiple documents that match the filter, only the first matching document is deleted.

delete_many() - This method is used to delete multiple documents from a collection that match the specified filter. All documents that match the filter are deleted.

drop() - This method is used to drop the entire collection from the database. All documents and indexes in the collection are deleted.
```
